```python
# Split natural language into sentence tokens
# Create a dataframe of the allowed combinations of sentences

from nltk.tokenize import sent_tokenize 
  
f = open("in/Nixon_debate.txt", "r")
text = f.read()
tokens = sent_tokenize(text)
```

```python
tokens
```

```python
tokens = ['A', 'B', 'C']
```

```python
pairs = []
 
# Cartesian product of arguments
for i in range(0,len(tokens)):
    for j in range(0,len(tokens)):
       
        if (i!=j):
            pairs.append((tokens[i],tokens[j]))
```

```python
import pandas as pd
df = pd.DataFrame(data=pairs)
df
```

```python
df.to_csv('out/pairs_test_nixon.csv', index=False)
```
